## 예제 문제
- 정규분포를 가지는 확률변수의 분산 σ2 = 1, 확률변수의 표본은 한개 x1 = 1은 알고 있으나 평균 μ를 모르고 있어 이를 추정해야 하는 문제
- 이 경우 어떤 μ 값이 가장 가능성(가능도)이 큰지
    - μ=−1, μ=0, μ=1, 세 가지 후보를 제시
    - 이 세 가지 μ 값에 대해 1이 나올 확률밀도의 값이 바로 가능도

- $N(x;μ=−1)$ 확률분포에서 x=1이 나올 가능도(확률밀도)는 0.05

- $N(x;μ=0)$ 확률분포에서 x=1이 나올 가능도(확률밀도)는 0.24

- $N(x;μ=1)$ 확률분포에서 x=1이 나올 가능도(확률밀도)는 0.40

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np

# x data 생성
x = np.linspace(-5, 5, 100)

p1 = stats.norm(loc=-1).pdf(1)
p2 = stats.norm(loc=0).pdf(1)
p3 = stats.norm(loc=1).pdf(1)

plt.scatter(1, p1, s=100, c='r', marker='v', 
         label=r"$N(x_1;\mu=-1)$={:.2f}".format(np.round(p1, 2)))
plt.scatter(1, p2, s=100, c='b', marker='^', 
         label=r"$N(x_1;\mu=0)$={:.2f}".format(np.round(p2, 2)))
plt.scatter(1, p3, s=100, c='g', marker='s', 
         label=r"$N(x_1;\mu=1)$={:.2f}".format(np.round(p3, 2)))

plt.plot(x, sp.stats.norm(loc=-1).pdf(x), ls="-.")
plt.plot(x, sp.stats.norm(loc=0).pdf(x), ls="--")
plt.plot(x, sp.stats.norm(loc=1).pdf(x), ls="-")

# 1개의 표본
plt.scatter(1, 0, s=100, c='k')
plt.vlines(1, -0.09, 0.45, linestyle=":")
plt.text(1-0.3, -0.15, "$x_1=1$")

# graph draw
plt.xlabel("x")
plt.ylabel("PDF")
plt.legend()
plt.title("Maximum likelihood")
plt.show()
```
